##  Exercice 5 – Compétences d'observation : Jeu de données Sommeil

**Objectif :** Identifier les colonnes intéressantes pour différents types d'analyses.

```python
print('=== Analyse exploratoire – Jeu de données Sommeil ===')
print('\nAperçu général :')
display(sleep_data.head())

print('\n--- Statistiques descriptives ---')
display(sleep_data.describe())

print('\n--- Distribution de la qualité du sommeil ---')
print(sleep_data['Sleep_Quality'].value_counts())
```

```python
# ── Colonne 1 : Hours_Sleep_Weekday – Analyse de tendance ──
# Intéressant pour observer si les Américains dorment suffisamment
# Recommandation OMS : 7 à 9 heures pour les adultes

avg_sleep = sleep_data['Hours_Sleep_Weekday'].mean()
median_sleep = sleep_data['Hours_Sleep_Weekday'].median()
below_recommended = (sleep_data['Hours_Sleep_Weekday'] < 7).sum()

print('── Analyse : Hours_Sleep_Weekday (Analyse de tendance) ──')
print(f'  Moyenne : {avg_sleep:.2f} heures')
print(f'  Médiane : {median_sleep:.2f} heures')
print(f'  Personnes dormant moins de 7h (sous le seuil OMS) : {below_recommended}/{len(sleep_data)}')
print(f'  → Pertinent pour analyser la tendance générale au manque de sommeil')

# ── Colonne 2 : Work_Hours_Per_Week vs Hours_Sleep_Weekday – Comparaison de groupes ──
print('\n── Analyse : Work_Hours_Per_Week vs Sommeil (Comparaison de groupes) ──')
# Créer des groupes de travailleurs
sleep_data['Work_Group'] = pd.cut(
    sleep_data['Work_Hours_Per_Week'],
    bins=[0, 40, 50, 100],
    labels=['Temps normal (≤40h)', 'Heures sup (41–50h)', 'Surcharge (>50h)']
)
comparison = sleep_data.groupby('Work_Group')['Hours_Sleep_Weekday'].mean().round(2)
print(comparison.to_string())
print(f'  → Pertinent pour comparer le sommeil selon l\'intensité de travail')

# ── Colonne 3 : Sleep_Quality – Distribution catégorielle ──
print('\n── Analyse : Sleep_Quality (Distribution catégorielle) ──')
quality_dist = sleep_data['Sleep_Quality'].value_counts()
print(quality_dist.to_string())
print(f'  → Pertinent pour comprendre la perception subjective du sommeil')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Observations clés – Données de Sommeil', fontsize=13, fontweight='bold')

# Graphique 1 : Distribution heures de sommeil
axes[0].hist(sleep_data['Hours_Sleep_Weekday'], bins=8, color='steelblue', edgecolor='white')
axes[0].axvline(7, color='red', linestyle='--', label='Seuil OMS (7h)')
axes[0].set_title('Heures de sommeil (semaine)')
axes[0].set_xlabel('Heures')
axes[0].set_ylabel('Fréquence')
axes[0].legend()

# Graphique 2 : Sommeil par groupe de travail
work_means = sleep_data.groupby('Work_Group')['Hours_Sleep_Weekday'].mean()
axes[1].bar(range(len(work_means)), work_means.values, color=['#4CAF50','#FF9800','#f44336'])
axes[1].set_xticks(range(len(work_means)))
axes[1].set_xticklabels(['≤40h', '41-50h', '>50h'], rotation=15)
axes[1].axhline(7, color='red', linestyle='--', alpha=0.7)
axes[1].set_title('Sommeil vs Heures de travail')
axes[1].set_ylabel('Heures de sommeil moy.')

# Graphique 3 : Distribution qualité du sommeil
quality_order = ['Poor', 'Fair', 'Good', 'Excellent']
quality_counts = sleep_data['Sleep_Quality'].value_counts().reindex(quality_order, fill_value=0)
colors = ['#f44336', '#FF9800', '#4CAF50', '#2196F3']
axes[2].bar(quality_order, quality_counts.values, color=colors, edgecolor='white')
axes[2].set_title('Qualité du sommeil')
axes[2].set_xlabel('Qualité')
axes[2].set_ylabel('Nombre de personnes')

plt.tight_layout()
plt.savefig('sleep_analysis.png', dpi=120, bbox_inches='tight')
plt.show()
print(' Graphique sauvegardé : sleep_analysis.png')
```

##  Exercice 3 (suite) – Cahier d'analyse : Statistiques et Visualisations

**Objectif :** Calculer moyenne, médiane, mode et créer des visualisations sur le jeu Iris.

```python
print('=== Statistiques descriptives complètes – Iris Dataset ===')
print('\n' + '='*60)

for col in quantitative_cols:
    mean_val   = iris_df[col].mean()
    median_val = iris_df[col].median()
    mode_val   = iris_df[col].mode()[0]
    std_val    = iris_df[col].std()
    min_val    = iris_df[col].min()
    max_val    = iris_df[col].max()
    
    print(f'\n📏 {col.replace("_", " ").upper()}')
    print(f'   Moyenne  : {mean_val:.3f} cm')
    print(f'   Médiane  : {median_val:.3f} cm')
    print(f'   Mode     : {mode_val:.3f} cm')
    print(f'   Écart-type : {std_val:.3f}')
    print(f'   Min – Max  : {min_val:.1f} – {max_val:.1f} cm')
```

```python
# ── Histogrammes avec courbe de densité ──

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Distributions des mesures Iris (avec densité)', fontsize=13, fontweight='bold')

colors = ['#3498db', '#e74c3c', '#2ecc71', '#9b59b6']

for idx, (col, color) in enumerate(zip(quantitative_cols, colors)):
    ax = axes[idx // 2][idx % 2]
    sns.histplot(iris_df[col], kde=True, color=color, ax=ax, bins=20, alpha=0.7)
    mean_v = iris_df[col].mean()
    median_v = iris_df[col].median()
    ax.axvline(mean_v, color='black', linestyle='--', linewidth=1.5, label=f'Moy: {mean_v:.2f}')
    ax.axvline(median_v, color='red', linestyle=':', linewidth=1.5, label=f'Med: {median_v:.2f}')
    ax.set_title(col.replace('_', ' ').title())
    ax.set_xlabel('Valeur (cm)')
    ax.legend(fontsize=9)

plt.tight_layout()
plt.savefig('iris_histograms.png', dpi=120, bbox_inches='tight')
plt.show()
print(' Graphique sauvegardé : iris_histograms.png')
```

```python
# ── Diagrammes à barres – Moyennes par espèce ──

iris_means = iris_df.groupby('species')[quantitative_cols].mean()

fig, ax = plt.subplots(figsize=(11, 6))
x = np.arange(len(iris_means.index))
width = 0.2
palette = ['#3498db', '#e74c3c', '#2ecc71', '#9b59b6']

for i, (col, color) in enumerate(zip(quantitative_cols, palette)):
    bars = ax.bar(x + i * width, iris_means[col], width,
                  label=col.replace('_', ' ').title(), color=color, alpha=0.85)
    # Ajouter les valeurs sur chaque barre
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05,
                f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=8)

ax.set_xlabel('Espèce')
ax.set_ylabel('Moyenne (cm)')
ax.set_title('Moyennes des mesures par espèce – Iris Dataset')
ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(iris_means.index)
ax.legend()

plt.tight_layout()
plt.savefig('iris_bar_chart.png', dpi=120, bbox_inches='tight')
plt.show()
print(' Graphique sauvegardé : iris_bar_chart.png')
```

```python
# ── Pairplot – Relations entre toutes les variables quantitatives ──

fig = sns.pairplot(iris_df, hue='species', palette='Set2', 
                   plot_kws={'alpha': 0.6}, diag_kind='kde')
fig.figure.suptitle('Pairplot – Relations entre variables Iris', y=1.02, fontsize=13)
plt.savefig('iris_pairplot.png', dpi=100, bbox_inches='tight')
plt.show()
print(' Graphique sauvegardé : iris_pairplot.png')
```

##  Exercice 10 – Lecture des données JSON depuis une URL

**Objectif :** Importer des données JSON depuis une URL et afficher les 5 premières entrées.

```python
# URL publique de données JSON – JSONPlaceholder (API de test populaire)
# Données : liste d'utilisateurs fictifs avec informations complètes

json_url = 'https://jsonplaceholder.typicode.com/users'

print(f' Chargement des données JSON depuis : {json_url}')

# Méthode 1 : Via Pandas directement
json_df = pd.read_json(json_url)

print(f' Données chargées avec succès !')
print(f'   Dimensions : {json_df.shape[0]} lignes × {json_df.shape[1]} colonnes')
print(f'\n--- Colonnes disponibles ---')
print(list(json_df.columns))
```

```python
# Affichage des 5 premières entrées

print('--- 5 premières entrées ---')
display(json_df.head(5))

# Extraction des colonnes simples pour un affichage plus lisible
simple_cols = ['id', 'name', 'username', 'email', 'phone', 'website']
available_cols = [c for c in simple_cols if c in json_df.columns]

print('\n--- Vue simplifiée (colonnes principales) ---')
display(json_df[available_cols].head(5))
```

```python
# Méthode 2 : Via requests + json (approche alternative explicative)

print('=== Méthode alternative : requests + json.loads ===')

response = requests.get(json_url)

if response.status_code == 200:
    data_raw = response.json()  # Convertit la réponse JSON en liste Python
    df_from_requests = pd.DataFrame(data_raw)
    
    print(f' Statut HTTP : {response.status_code} OK')
    print(f'   Nombre d\'entrées : {len(data_raw)}')
    print(f'\n--- 5 premières entrées via requests ---')
    display(df_from_requests[available_cols].head(5))
else:
    print(f' Erreur HTTP : {response.status_code}')
```
