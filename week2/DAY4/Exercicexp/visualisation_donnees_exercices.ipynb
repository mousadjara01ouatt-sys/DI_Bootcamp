## Exercice 1 : Comprendre la visualisation des données

### Importance de la visualisation des données
La visualisation des données transforme des chiffres abstraits en éléments visuels concrets. Elle est cruciale car elle permet d'identifier instantanément des tendances globales, de repérer des anomalies ou valeurs aberrantes (outliers) et de communiquer efficacement des analyses complexes à un public non technique.

### Objectif d'un graphique linéaire
L'objectif principal d'un graphique linéaire est d'illustrer des relations continues, le plus souvent pour suivre des tendances, des cycles ou des variations de données sur une période temporelle séquentielle.

## Exercice 2 : Graphique linéaire pour la variation de température

```python
import matplotlib.pyplot as plt

jours = ['Lun', 'Mar', 'Mer', 'Jeu', 'Ven', 'Sam', 'Dim']
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.figure(figsize=(8, 4))
plt.plot(jours, temperatures, marker='o', color='orange', linewidth=2)
plt.title("Variations de la température sur une semaine")
plt.xlabel("Jour")
plt.ylabel("Température (°F)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## Exercice 3 : Graphique à barres pour les ventes mensuelles

```python
import matplotlib.pyplot as plt

mois = ['Janv', 'Févr', 'Mars', 'Avril', 'Mai']
ventes = [5000, 5500, 6200, 7000, 7500]

plt.figure(figsize=(8, 5))
plt.bar(mois, ventes, color='skyblue', edgecolor='royalblue')
plt.title("Montant des ventes mensuelles")
plt.xlabel("Mois")
plt.ylabel("Montant des ventes ($)")
plt.show()
```

## Initialisation du Jeu de Données : Santé mentale des étudiants


```python
import pandas as pd
import numpy as np
import seaborn as sns

np.random.seed(42)
donnees_simulees = {
    'Choose your gender': np.random.choice(['Male', 'Female'], 100),
    'Age': np.random.randint(18, 26, 100),
    'What is your CGPA?': np.random.choice(['0 - 1.99', '2.00 - 2.49', '2.50 - 2.99', '3.00 - 3.49', '3.50 - 4.00'], 100),
    'Do you have Anxiety?': np.random.choice(['Yes', 'No'], 100),
    'Do you have Panic Attacks?': np.random.choice(['Yes', 'No'], 100)
}
df = pd.DataFrame(donnees_simulees)
```

## Exercise 4: Visualizing the Distribution of CGPA

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9, 5))
sns.histplot(data=df, x='What is your CGPA?', color='purple', shrink=0.8)
plt.title("Distribution du CGPA des étudiants")
plt.xlabel("Tranches de CGPA")
plt.ylabel("Nombre d'étudiants")
plt.show()
```

## Exercise 5: Comparing Anxiety Levels Across Different Genders

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['Anxiety_Numeric'] = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})

plt.figure(figsize=(7, 5))
sns.barplot(data=df, x='Choose your gender', y='Anxiety_Numeric', palette='pastel', errorbar=None)
plt.title("Proportion d'étudiants souffrant d'anxiété selon le genre")
plt.xlabel("Genre")
plt.ylabel("Proportion d'anxiété (0 à 1)")
plt.show()
```

## Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['Panic_Attacks_Numeric'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='Age',
    y='Panic_Attacks_Numeric',
    s=120,
    color='crimson',
    alpha=0.7,
    marker='X'
)
plt.title("Relation entre l'âge et la présence de crises de panique")
plt.xlabel("Âge des étudiants")
plt.ylabel("Crise de panique (1 = Oui, 0 = Non)")
plt.yticks([0, 1], ['Non (0)', 'Oui (1)'])
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```
