# 🏬 Mini-Projet : Analyse des données pour la stratégie marketing (US Superstore)
**Objectif :** Analyser le véritable jeu de données `superstore_dataset.csv` pour en extraire des insights marketing, identifier les clients clés à l'aide de la loi de Pareto, et répondre précisément aux questions stratégiques sans altération artificielle des données.

## Étape 1 : Chargement et Prétraitement des Données
Nous configurons notre environnement et chargeons le fichier authentique demandé afin de mener des analyses basées sur la réalité des transactions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12

# Chargement du jeu de données réel requis
try:
    df = pd.read_csv('superstore_dataset.csv', encoding='ISO-8859-1')
    print("✅ Le jeu de données 'superstore_dataset.csv' a été chargé avec succès !")
except FileNotFoundError:
    print("❌ Fichier 'superstore_dataset.csv' introuvable. Veuillez vérifier son emplacement.")

print(f"Dimensions du jeu de données : {df.shape[0]} lignes, {df.shape[1]} colonnes")
```

### Nettoyage et conversion des types
Nous harmonisons les types de données temporelles et supprimons les doublons éventuels.

```python
# Suppression des doublons
df = df.drop_duplicates()

# Conversion des colonnes de dates
for col in ['Order Date', 'Ship Date']:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col])

print("Aperçu des colonnes et types mis à jour :")
df.info()
```

## Étape 2 : Analyse des Clients Importants et Loi de Pareto

### Question : Quels sont les 20 clients les plus importants selon les ventes ?
Pour répondre précisément à cette question sans biais ni injection artificielle de données (comme la création fictive de 'Jean Dupont'), nous agrégeons le montant total des ventes par client présent naturellement dans la base de données.

```python
# Détermination de la colonne d'identification du client (ex: 'Customer Name' ou 'Customer ID')
cust_col = 'Customer Name' if 'Customer Name' in df.columns else ('Customer ID' if 'Customer ID' in df.columns else df.columns[0])

# Agrégation des ventes par client
client_sales = df.groupby(cust_col)['Sales'].sum().reset_index()
client_sales = client_sales.sort_values(by='Sales', ascending=False).reset_index(drop=True)

print("📊 TABLEAU DES 20 PREMIERS CLIENTS PAR CHIFFRE D'AFFAIRES :")
top_20_clients = client_sales.head(20)
display(top_20_clients)
```

### Visualisation graphique des 20 premiers clients

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=top_20_clients, x='Sales', y=cust_col, palette='viridis')
plt.title("Top 20 des clients les plus importants selon le volume de ventes")
plt.xlabel("Total des Ventes ($)")
plt.ylabel("Nom du Client")
plt.tight_layout()
plt.show()
```

### Application de la Loi de Pareto (Règle des 80/20)
Nous calculons le pourcentage cumulé des ventes par rapport au pourcentage cumulé des clients pour vérifier si une minorité de clients génère la majorité du chiffre d'affaires.

```python
client_sales['Sales_Cum_Pct'] = (client_sales['Sales'].cumsum() / client_sales['Sales'].sum()) * 100
client_sales['Clients_Cum_Pct'] = (np.arange(1, len(client_sales) + 1) / len(client_sales)) * 100

# Trouver la valeur exacte des ventes cumulées pour 20% des clients
idx_20 = (client_sales['Clients_Cum_Pct'] - 20).abs().idxmin()
sales_at_20 = client_sales.loc[idx_20, 'Sales_Cum_Pct']

plt.figure(figsize=(10, 6))
plt.plot(client_sales['Clients_Cum_Pct'], client_sales['Sales_Cum_Pct'], color='purple', lw=2.5, label='Ventes cumulées')
plt.axvline(x=20, color='red', linestyle='--', alpha=0.7, label='Seuil 20% des clients')
plt.axhline(y=sales_at_20, color='grey', linestyle='--', alpha=0.7, label=f'{sales_at_20:.1f}% des ventes')
plt.title(f"Courbe de Pareto appliquée aux ventes\n(20% des clients représentent {sales_at_20:.2f}% du CA)")
plt.xlabel('% Cumulé des Clients')
plt.ylabel('% Cumulé des Ventes')
plt.legend()
plt.show()
```
