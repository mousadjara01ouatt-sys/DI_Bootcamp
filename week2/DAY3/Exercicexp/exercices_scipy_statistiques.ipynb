## Exercice 2 : Statistiques descriptives
**Tâche :** À partir d’un échantillon de données, calculez la moyenne, la médiane, la variance et l’écart-type à l’aide de SciPy.

Jeu de données exemple : `[12, 15, 13, 12, 18, 20, 22, 21]`

```python
import numpy as np
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

moyenne = np.mean(data)
mediane = np.median(data)
variance = stats.tvar(data)  # Variance échantillonnale (degré de liberté = 1)
ecart_type = stats.tstd(data)  # Écart-type échantillonnal (degré de liberté = 1)

print(f"Moyenne : {moyenne}")
print(f"Médiane : {mediane}")
print(f"Variance (échantillon) : {variance:.2f}")
print(f"Écart-type (échantillon) : {ecart_type:.2f}")
```

## Exercice 3 : Comprendre les distributions
**Tâche :** Générer une distribution normale en utilisant SciPy avec une moyenne de 50 et un écart-type de 10. Tracez cette distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 50
sigma = 10

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Normale(mu={mu}, sigma={sigma})', color='blue')
plt.title('Distribution Normale')
plt.xlabel('Valeurs')
plt.ylabel('Densité de probabilité')
plt.legend()
plt.grid(True)
plt.show()
```

## Exercice 4 : Application du test T
**Tâche :** Effectuer un test T sur deux ensembles de données générées aléatoirement.

```python
import numpy as np
from scipy.stats import ttest_ind

np.random.seed(42)
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

statistique_t, p_valeur = ttest_ind(data1, data2)

print(f"Statistique T : {statistique_t:.4f}")
print(f"P-valeur : {p_valeur:.4e}")

alpha = 0.05
if p_valeur < alpha:
    print("Rejet de l'hypothèse nulle : les moyennes sont significativement différentes.")
else:
    print("Échec du rejet de l'hypothèse nulle : pas de différence significative.")
```

## Exercise 5: Linear Regression Analysis
**Task:** Given a dataset of housing prices and their corresponding sizes, use linear regression to predict the price of a house given its size.

### Questions & Réponses
* **Pente et ordonnée à l'origine :** La pente indique la variation moyenne du prix par mètre carré supplémentaire. L'ordonnée à l'origine est la valeur théorique pour 0 m².
* **Prédiction pour 90 m² :** Calculée via la formule de la droite de régression : $y = ax + b$.
* **Interprétation de la pente :** La valeur représente l'augmentation du prix de la maison pour chaque unité de surface supplémentaire.

```python
import numpy as np
from scipy.stats import linregress

house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

slope, intercept, r_value, p_value, std_err = linregress(house_sizes, house_prices)

predicted_size = 90
predicted_price = slope * predicted_size + intercept

print(f"Pente (Slope) : {slope:.2f}")
print(f"Ordonnée à l'origine (Intercept) : {intercept:.2f}")
print(f"Prix prédit pour une maison de 90 m² : {predicted_price:.2f}")
```

## Exercise 6: Understanding ANOVA
**Task:** Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

### Questions & Réponses
* **Valeur F et P-valeur :** Fournies par le test `f_oneway`.
* **Effet des engrais :** Si la P-valeur est inférieure à 0.05, l'effet est considéré comme significativement différent.
* **Si la P-valeur était supérieure à 0.05 :** Nous ne pourrions pas affirmer que les engrais ont un effet différent; les écarts observés pourraient être dus au simple hasard.

```python
from scipy.stats import f_oneway

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

f_val, p_val = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"Valeur F (F-value) : {f_val:.4f}")
print(f"P-valeur (P-value) : {p_val:.4e}")
```

## Exercise 7: Probability Distributions (Optional)
**Task:** Calculate the probability of getting exactly 5 heads in 10 coin flips using a binomial distribution.

```python
from scipy.stats import binom

n = 10
p = 0.5
succes = 5

probabilite = binom.pmf(succes, n, p)
print(f"La probabilité d'obtenir exactement 5 piles en 10 lancers est de : {probabilite:.4f} ({probabilite*100:.2f}%)")
```

## Exercice 8 : Coefficients de corrélation (optionnel)
**Tâche :** Calculez les coefficients de corrélation de Pearson et Spearman entre deux variables (âge et revenus).

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

corr_pearson, p_pearson = pearsonr(data['age'], data['income'])
corr_spearman, p_spearman = spearmanr(data['age'], data['income'])

print(f"Coefficient de Pearson : {corr_pearson:.4f} (p-valeur: {p_pearson:.4f})")
print(f"Coefficient de Spearman : {corr_spearman:.4f} (p-valeur: {p_spearman:.4f})")
```
